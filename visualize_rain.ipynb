```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Data
y_values = [611, 62, 2478, 138, 925, 1116, 647]
x_labels = ["2017-08-07 to 2017-08-17", "2018-09-02 to 2018-09-12",
            "2019-07-05 to 2019-07-15", "2020-09-15 to 2020-09-25",
            "2021-10-15 to 2021-10-22", "2022-06-20 to 2022-06-30",
            "2023-07-07 to 2023-07-17"]

x = np.arange(len(x_labels))

# Create bar diagram
plt.figure(figsize=(10, 6))
bars = plt.bar(x, y_values, color='red', alpha=0.8, edgecolor = 'black')

# Add values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 20, int(yval), ha='center', va='bottom', fontsize=9)

# Customize plot
plt.xlabel('Flood Period')
plt.ylabel('Flood Area (Ha)')
plt.title('YEAR BY YEAR FLOOD AFFECTED AREA')

# Shift xticks towards left by adding a small offset to x
offset = -0.2
plt.xticks(x + offset, x_labels, rotation=45)

plt.ylim(0, max(y_values) + 200)  # Set y-axis limits

# Show plot
plt.tight_layout()
plt.show()
```
